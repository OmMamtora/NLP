# Text Cleaning

### Removing HTML Tags

```python
sample_text = '<h2>HTML Element</h2><p>The HTML <code class="w3-codespan">&lt;sup&gt ;< /code> element defines superscript text. Superscript text appears half a character above the normal line, and is sometimes rendered in a smaller font. Superscript text can be used for footnotes, like WWW<sup>[1]</s up> : </p>'
```

```python
sample_text
```

```python
import re

def stripehtml(data):
    p = re.compile(r'<.*?>')
    return p.sub('', data)
```

```python
stripehtml(sample_text)
```

#### Word and Sentence Tokenizer

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
sents= sent_tokenize(dummy)
print(sents)

for sentences in sents:
    print(sentences)
```

```python
words = word_tokenize(dummy)
print(words)
```

```python
# Word tokenize based on sentences

for sentences in sents:
    print(word_tokenize(sentences))
```

```python
from textblob import TextBlob

def text_correct():
    data = input("Enter Sentence :->")
    text_blob = TextBlob(data)
    correct = text_blob.correct()

    print("Correct Sentence :-> ", correct)
```
