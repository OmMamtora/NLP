```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.DataFrame({
    'text' : ['people watch campusx','campusx watch campusx', 'people write comment','campusx write comment'],
    'output' : [1,1,0,0]
})

df
```

```python
cv = CountVectorizer()
bow = cv.fit_transform(df['text'])
```

```python
# Vocab

print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```
