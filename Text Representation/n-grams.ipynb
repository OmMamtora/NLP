```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.DataFrame({
    'text' : ['people watch campusx','campusx watch campusx', 'people write comment','campusx write comment'],
    'output' : [1,1,0,0]
})

df
```

#### Default word selection is 1 and it is called unigram OR Bag of Word

```python
cv = CountVectorizer()
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
# Bi-Grams

cv1 = CountVectorizer(ngram_range=(2,2))
bow1 = cv1.fit_transform(df['text'])
```

```python
print(cv1.vocabulary_)
```

```python
# Uni and Bi-Grams

cv2 = CountVectorizer(ngram_range=(1,2))
bow2 = cv2.fit_transform(df['text'])
```

```python
print(cv2.vocabulary_)
```

```python
# Tri-Grams

cv3 = CountVectorizer(ngram_range=(3,3))
bow3 = cv3.fit_transform(df['text'])
```

```python
print(cv3.vocabulary_)
```
